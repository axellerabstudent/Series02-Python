## Sine Wave

$$
y(t) = A \,\sin(\omega t + \phi)
$$


Where:  
- \( A \): amplitude  
- \( \omega \): angular frequency (\( \omega = 2 \pi f \))  
- \( t \): time  
- \( \phi \): phase


```python
import numpy as np
import matplotlib.pyplot as plt

# Time array (1000 values from 0 to 1)
t = np.linspace(0, 1, 1000)

# Sine signal
s = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave (you can change the frequency)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(t, s, label="Sine wave")
plt.title("Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

# Save as PNG
plt.savefig("sine_wave.png", dpi=300)

# Show the plot
plt.show()
```

```python
# Plot (showing only one period)
plt.figure(figsize=(8, 4))
plt.plot(t, s, label="Sine wave")
plt.title("One Period of a 5 Hz Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

# Limit x-axis to one period (0 → 0.2s for 5 Hz)
plt.xlim(0, 1/5)

# Save as PNG
plt.savefig("sine_wave_one_period.png", dpi=300)

# Show the plot
plt.show()
```

```python
# Time array (1000 values from 0 to 1)
t = np.linspace(0, 1, 1000)

# Sine signal
s = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave (you can change the frequency)

# Cosine signal
c = np.cos(2 * np.pi * 5 * t)  # 5 Hz cosine wave

# Plot
plt.figure(figsize=(8, 4))
plt.plot(t, s, label="Sine wave")
plt.plot(t, c, label="Cosine wave")
plt.title("Sine and Cosine Waves")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

# Limit x-axis to one period (0 → 0.2s for 5 Hz)
plt.xlim(0, 1/5)

# Save as PNG
plt.savefig("sine_cosine_wave.png", dpi=300)

# Show the plot
plt.show()
```
